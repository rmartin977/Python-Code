# Gradient Ascent


Python program to find the maximum of the function:
 $$f(x,y) = 10-x^2-2y^2$$ using gradient ascent.

@Robert Martin
updated 5/24/21



## Import **standard** libraries

```
import numpy as np
import matplotlib.pyplot as plt
```

## Define the function and partial derivatives



```
def f(x,y):
  return 10 - x**2 - 2*y**2
def dx(x):
  return -2*x
def dy(y):
  return -4*y
```

##Intialize starting values to x = 4 and y = 4

```
x,y = (-4,4)
alpha = .1  # 'learning rate
x_data = [x]
y_data = [y]
```

```
x_data[:5]
```

## Perform Gradient Ascent

```
for k in range(50):
  x = x + alpha*dx(x)
  y = y + alpha *dy(y)
  x_data.append(x)
  y_data.append(y)
```

```
x_data[:10]
```

##Plot the results

```
xx = np.linspace(-6,6,100)
yy = np.linspace(-6,6,100)
X,Y = np.meshgrid(xx,yy)
Z = f(X,Y)

           
```

```
plt.figure(figsize=(10,10))
plt.plot(x_data,y_data,'.-')
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.grid()
plt.title('Gradient Ascent');
plt.contour(X,Y,Z);
```

```
print(f'The solution is {(x_data[-1],y_data[-1])}')
```
