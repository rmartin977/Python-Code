<h1> Gradient Descent Example </h1>

## Import standard libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define the function and partial derivatives



```python
def f(x,y):
  return -(10 - x**2 - y**2)
def dx(x):
  return 2*x
def dy(y):
  return 2*y
```

## Intialize starting values to x = 4 and y = 4

```python
x,y = (4,4)
alpha = .05  # 'learning rate
x_data = [x]
y_data = [y]
```

```python
x_data
```

## Perform Gradient Descent

```python
for k in range(500):
  x = x - alpha*dx(x)
  y = y - alpha *dy(y)
  x_data.append(x)
  y_data.append(y)
```

## Plot the results

```python
# xx = np.linspace(-6,6,100)
# yy = np.linspace(-6,6,100)
# X,Y = np.meshgrid(xx,yy)
# Z = f(X,Y)

           
```

```python
xx = np.linspace(-6,6,100)
yy = np.linspace(-6,6,100)
X,Y = np.meshgrid(xx,yy)
Z = f(X,Y)

      
```

```python
x_data;
```

```python
plt.figure(figsize=(10,10))
plt.plot(x_data,y_data,'.-')
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.grid()
plt.title('Gradient Descent');
plt.contour(X,Y,Z);
plt.axis("equal")
```
